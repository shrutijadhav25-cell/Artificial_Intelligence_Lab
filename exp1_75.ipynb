```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
num1 = 121
print(is_palindrome(num1))

num2 = 12321
print(is_palindrome(num2))

num3 = 12345
print(is_palindrome(num3))

num4 = 0
print(is_palindrome(num4))
```

## Define Decision Rules

### Subtask:
Outline a set of simple decision rules for an expert system that recommends activities based on weather conditions.


### Decision Rules for Activity Recommendation System

To outline simple decision rules for an expert system recommending activities based on weather conditions, we first identify primary weather conditions and then define 'if-then' rules for various activities.

#### 1. Primary Weather Conditions:

*   **Temperature:**
    *   `Hot`: > 28°C
    *   `Warm`: 20°C - 28°C
    *   `Mild`: 10°C - 19°C
    *   `Cold`: < 10°C
*   **Precipitation:**
    *   `None`
    *   `Rain`
    *   `Snow`
*   **Wind Speed:**
    *   `Calm`: < 10 km/h
    *   `Breezy`: 10 km/h - 25 km/h
    *   `Windy`: > 25 km/h
*   **Cloud Cover:**
    *   `Clear`
    *   `Partly Cloudy`
    *   `Overcast`

#### 2. Activity Recommendations and Decision Rules (If-Then Statements):

*   **Activity: Swimming**
    *   **Rule 1:** IF Temperature is `Hot` AND Precipitation is `None` THEN Recommend `Swimming`.
    *   **Rule 2:** IF Temperature is `Warm` AND Precipitation is `None` AND Cloud Cover is `Clear` OR `Partly Cloudy` THEN Recommend `Swimming`.

*   **Activity: Hiking**
    *   **Rule 1:** IF Temperature is `Mild` AND Precipitation is `None` AND Wind Speed is `Calm` OR `Breezy` THEN Recommend `Hiking`.
    *   **Rule 2:** IF Temperature is `Warm` AND Precipitation is `None` AND Wind Speed is `Calm` AND Cloud Cover is `Overcast` THEN Recommend `Hiking`.

*   **Activity: Reading Indoors**
    *   **Rule 1:** IF Precipitation is `Rain` OR `Snow` THEN Recommend `Reading Indoors`.
    *   **Rule 2:** IF Temperature is `Cold` AND Wind Speed is `Windy` THEN Recommend `Reading Indoors`.

*   **Activity: Skiing**
    *   **Rule 1:** IF Precipitation is `Snow` AND Temperature is `Cold` THEN Recommend `Skiing`.

*   **Activity: Cycling**
    *   **Rule 1:** IF Temperature is `Mild` AND Precipitation is `None` AND Wind Speed is `Calm` THEN Recommend `Cycling`.

*   **Activity: Picnic**
    *   **Rule 1:** IF Temperature is `Warm` AND Precipitation is `None` AND Cloud Cover is `Clear` OR `Partly Cloudy` AND Wind Speed is `Calm` THEN Recommend `Picnic`.


## Implement Expert System Core

### Subtask:
Write Python code to implement the expert system based on the defined decision rules. This will include functions to categorize weather conditions and an inference engine to process user inputs and make decisions.


**Reasoning**:
I will define functions to categorize weather parameters based on the instructions, which will be used by the main inference engine function.



```python
def get_temperature_category(temp_celsius):
    if temp_celsius >= 25:
        return "Hot"
    elif 15 <= temp_celsius < 25:
        return "Mild"
    elif 0 <= temp_celsius < 15:
        return "Cool"
    else:
        return "Cold"

def get_precipitation_category(precipitation_type):
    if precipitation_type.lower() == "heavy rain" or precipitation_type.lower() == "snow":
        return "Heavy Precipitation"
    elif precipitation_type.lower() == "light rain" or precipitation_type.lower() == "drizzle":
        return "Light Precipitation"
    else:
        return "No Precipitation"

def get_wind_speed_category(wind_speed_kmh):
    if wind_speed_kmh >= 40:
        return "Strong Wind"
    elif 20 <= wind_speed_kmh < 40:
        return "Moderate Wind"
    else:
        return "Light Wind"

def get_cloud_cover_category(cloud_cover_percent):
    if cloud_cover_percent >= 75:
        return "Overcast"
    elif 25 <= cloud_cover_percent < 75:
        return "Partly Cloudy"
    else:
        return "Clear Skies"

print("Weather categorization functions defined.")
```

**Reasoning**:
I will create the `recommend_activity` function, which will serve as the expert system's inference engine, by implementing the decision rules using `if-elif-else` statements and utilizing the previously defined weather categorization functions. I will also make sure to use the specific categories defined in the 'Define Decision Rules' step, which differ slightly from the categories implemented in the previous step.



```python
def get_temperature_category_rules(temp_celsius):
    if temp_celsius > 28:
        return "Hot"
    elif 20 <= temp_celsius <= 28:
        return "Warm"
    elif 10 <= temp_celsius < 20:
        return "Mild"
    else:
        return "Cold"

def get_wind_speed_category_rules(wind_speed_kmh):
    if wind_speed_kmh < 10:
        return "Calm"
    elif 10 <= wind_speed_kmh <= 25:
        return "Breezy"
    else:
        return "Windy"

def recommend_activity(temperature, precipitation, wind_speed, cloud_cover):
    temp_cat = get_temperature_category_rules(temperature)
    wind_cat = get_wind_speed_category_rules(wind_speed)

    # Activity: Swimming
    if temp_cat == "Hot" and precipitation == "None":
        return "Swimming"
    if temp_cat == "Warm" and precipitation == "None" and (cloud_cover == "Clear" or cloud_cover == "Partly Cloudy"):
        return "Swimming"

    # Activity: Hiking
    if temp_cat == "Mild" and precipitation == "None" and (wind_cat == "Calm" or wind_cat == "Breezy"):
        return "Hiking"
    if temp_cat == "Warm" and precipitation == "None" and wind_cat == "Calm" and cloud_cover == "Overcast":
        return "Hiking"

    # Activity: Reading Indoors
    if precipitation == "Rain" or precipitation == "Snow":
        return "Reading Indoors"
    if temp_cat == "Cold" and wind_cat == "Windy":
        return "Reading Indoors"

    # Activity: Skiing
    if precipitation == "Snow" and temp_cat == "Cold":
        return "Skiing"

    # Activity: Cycling
    if temp_cat == "Mild" and precipitation == "None" and wind_cat == "Calm":
        return "Cycling"

    # Activity: Picnic
    if temp_cat == "Warm" and precipitation == "None" and (cloud_cover == "Clear" or cloud_cover == "Partly Cloudy") and wind_cat == "Calm":
        return "Picnic"

    return "No specific recommendation for these conditions"

print("Expert system inference engine defined.")

# Test cases to verify the rules (Optional, but good practice)
print(f"Hot, None, Calm, Clear: {recommend_activity(30, 'None', 5, 'Clear')}")
print(f"Warm, None, Breezy, Partly Cloudy: {recommend_activity(22, 'None', 15, 'Partly Cloudy')}")
print(f"Mild, None, Calm, Overcast: {recommend_activity(15, 'None', 5, 'Overcast')}")
print(f"Cold, Snow, Windy, Overcast: {recommend_activity(-5, 'Snow', 30, 'Overcast')}")
print(f"Rain, any, any, any: {recommend_activity(10, 'Rain', 10, 'Partly Cloudy')}")
print(f"No matching rule: {recommend_activity(5, 'None', 20, 'Overcast')}")
```

**Reasoning**:
The previous test run revealed that some test cases failed due to the order of evaluation of decision rules in the `recommend_activity` function. More specific rules were being overshadowed by broader rules. To fix this, I need to reorder the `if` statements in the `recommend_activity` function, placing more specific activity recommendations before more general ones that might partially overlap.



```python
def get_temperature_category_rules(temp_celsius):
    if temp_celsius > 28:
        return "Hot"
    elif 20 <= temp_celsius <= 28:
        return "Warm"
    elif 10 <= temp_celsius < 20:
        return "Mild"
    else:
        return "Cold"

def get_wind_speed_category_rules(wind_speed_kmh):
    if wind_speed_kmh < 10:
        return "Calm"
    elif 10 <= wind_speed_kmh <= 25:
        return "Breezy"
    else:
        return "Windy"

def recommend_activity(temperature, precipitation, wind_speed, cloud_cover):
    temp_cat = get_temperature_category_rules(temperature)
    wind_cat = get_wind_speed_category_rules(wind_speed)

    # Order matters: Specific rules should come before general ones

    # Activity: Skiing (Most specific in precipitation and temperature)
    if precipitation == "Snow" and temp_cat == "Cold":
        return "Skiing"

    # Activity: Picnic (Specific combination of Warm, No Precip, Clear/Partly Cloudy, and Calm Wind)
    if temp_cat == "Warm" and precipitation == "None" and (cloud_cover == "Clear" or cloud_cover == "Partly Cloudy") and wind_cat == "Calm":
        return "Picnic"

    # Activity: Cycling (Specific for Mild, No Precip, Calm Wind)
    if temp_cat == "Mild" and precipitation == "None" and wind_cat == "Calm":
        return "Cycling"

    # Activity: Swimming (More general than Picnic in Warm, No Precip, Clear/Partly Cloudy)
    if temp_cat == "Hot" and precipitation == "None":
        return "Swimming"
    if temp_cat == "Warm" and precipitation == "None" and (cloud_cover == "Clear" or cloud_cover == "Partly Cloudy"):
        return "Swimming"

    # Activity: Hiking (More general than Cycling in Mild, No Precip, Calm/Breezy Wind)
    if temp_cat == "Mild" and precipitation == "None" and (wind_cat == "Calm" or wind_cat == "Breezy"):
        return "Hiking"
    if temp_cat == "Warm" and precipitation == "None" and wind_cat == "Calm" and cloud_cover == "Overcast":
        return "Hiking"

    # Activity: Reading Indoors (General catch-all for bad weather, moved after more specific bad weather activities like Skiing)
    if precipitation == "Rain" or precipitation == "Snow":
        return "Reading Indoors"
    if temp_cat == "Cold" and wind_cat == "Windy":
        return "Reading Indoors"

    return "No specific recommendation for these conditions"

print("Expert system inference engine re-defined with rule priority adjustments.")

# Test cases to verify the rules (Optional, but good practice)
print(f"Hot, None, Calm, Clear: {recommend_activity(30, 'None', 5, 'Clear')}")
print(f"Warm, None, Breezy, Partly Cloudy: {recommend_activity(22, 'None', 15, 'Partly Cloudy')}")
print(f"Mild, None, Calm, Overcast: {recommend_activity(15, 'None', 5, 'Overcast')}")
print(f"Cold, Snow, Windy, Overcast: {recommend_activity(-5, 'Snow', 30, 'Overcast')}")
print(f"Rain, any, any, any: {recommend_activity(10, 'Rain', 10, 'Partly Cloudy')}")
print(f"No matching rule: {recommend_activity(5, 'None', 20, 'Overcast')}")
```
